```python
# POO
from abc import * 

class Evaluator(ABC):
    
    @abstractmethod
    def evaluate(self, val):
        pass
    
class Convertor(Evaluator):
    
    def __init__(self,convert_string):
        self._convert_string_ = convert_string
    
    def convert(self,val):
        if(self.evaluate(val)):
            return self._convert_string_
        return val
    
    
class NextConvertor(Convertor):
    
    def __init__(self,convert_string,next_convertor):
        Convertor.__init__(self,convert_string)
        self._next_convertor_ = next_convertor
    
    def convert(self,val):
        convert_result=super().convert(val)
        if(self._isNotSatisfy_(convert_result,val) and (self._has_next_())):
            return self._next_convertor_.convert(val)
        return convert_result
    
    def _has_next_(self):
        return (self._next_convertor_ != None)
    
    def _isNotSatisfy_(self,val1,val2):
        return (val1 == val2)

    
class NoConvertor(Convertor):
    
    def __init__(self):
        Convertor.__init__(self,"")
    
    def evaluate(self, val):
        return False
    
    def convert(self,val):
        return val
    
class Fizz(NextConvertor):
    
    def __init__(self,convert_string,next_convertor):
        NextConvertor.__init__(self,convert_string,next_convertor)
    
    def evaluate(self,val):
        return val%3 == 0
    
          
class Buzz(NextConvertor):
    
    def __init__(self,convert_string,next_convertor):
        NextConvertor.__init__(self,convert_string,next_convertor)
    
    def evaluate(self,val):
        return val%5 == 0

        
class FizzBuzz(NextConvertor):
    
    def __init__(self,convertString,fizz,buzz):
        NextConvertor.__init__(self,convertString,fizz)
        self._fizz_ = fizz
        self._buzz_ = buzz
    
    
    def evaluate(self,val):
        return self._fizz_.evaluate(val) & self._buzz_.evaluate(val)
```

```python
no=NoConvertor()
b=Buzz("Buzz",no)
f=Fizz("Fizz",b)
fb=FizzBuzz("FizzBuzz",f,b)

print([fb.convert(val) for val in range(100)])
```

```python
# Reactif
import rx
from rx import of, operators as op

isFizz = lambda val : (type(val) != str) and (val%3 == 0)
isBuzz = lambda val : (type(val) != str) and (val%5 == 0)
isFizzBuzz = lambda val : isFizz(val) & isBuzz(val)


of(*[val for val in range(100)]).pipe(
    op.map(lambda x : "FizzBuzz" if isFizzBuzz(x) else x),
    op.map(lambda x : "Buzz" if isBuzz(x) else x),
    op.map(lambda x : "Fizz" if isFizz(x) else x)).subscribe(lambda x : print(x))


```
