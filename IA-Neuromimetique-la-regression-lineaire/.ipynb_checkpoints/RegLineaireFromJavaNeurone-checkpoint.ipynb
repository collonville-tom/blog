```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import math

def matAbs(A,B):
    Z=[]
    for i in range(0,len(A)):
        if A[i]>B[i] :
            Z.append(A[i])
        else:
            Z.append(B[i])
    return np.array(Z)

def matlog(D):
    DAT=[]
    for i in D:
        if i!= 0:
            DAT.append(math.log(abs(i)))
        else :
            DAT.append(i)
    return np.array(DAT)
```

```python
learnSet=pd.read_csv('dataRegressionFromJava4/learnSet.csv',sep='\t')
maeDerEvol=pd.read_csv('dataRegressionFromJava4/maeDerEvol.csv',sep='\t')
maeEvol=pd.read_csv('dataRegressionFromJava4/maeEvol.csv',sep='\t')
maeMap=pd.read_csv('dataRegressionFromJava4/maeMap.csv',sep='\t')
print(learnSet[:3])
print(maeDerEvol[:3])
print(maeEvol[:3])
print(maeMap[:3])
```
