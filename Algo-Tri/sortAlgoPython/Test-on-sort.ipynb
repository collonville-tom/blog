```python
import time
from random import randrange
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

dataFrame=pd.read_pickle("python-sort.txt")
dataFrame
```

```python
print(dataFrame.keys())
for key in [ x for x in dataFrame.keys() ] :
    print(key)
```

```python
serie=dataFrame['bourrin']
print("Moyenne {}".format(np.mean(serie[10])))
print("Medianne {}".format(np.median(serie[10])))
print("Ecart type {}".format(np.std(serie[10])))
```

```python


resultDicMoy={}
resultDicEcart={}
#if x == 'system_sort' or x == 'quick_sort'
for key in [ x for x in dataFrame.keys() ] : 
    moy=[]
    std=[]
    for serkey in [ x for x in dataFrame[key].keys() ] :
        moy.append(np.mean(dataFrame[key][serkey]))
        std.append(np.std(dataFrame[key][serkey]))
    serieMoy=pd.Series(moy)
    resultDicMoy[key]=serieMoy
    resultDicEcart[key]=std

    
dataMoy=pd.DataFrame(resultDicMoy)
dataStd=pd.DataFrame(resultDicEcart)  

dataMoy.plot(figsize=(10,10))
dataStd.plot(figsize=(10,10))
```
